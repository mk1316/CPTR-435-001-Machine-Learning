# Pandas 4 - Statistics 

```python
import pandas as pd
```

```python
df = pd.read_csv(r'house-prices.csv', sep=',')
df
```

```python
# Mean
mean = df['Price'].mean()
print(mean)
# Median
median = df['Price'].median()
print(median)
# Mode
mode = df['Neighborhood'].mode()[0]
print(mode)
# Sum
sum_ = df['Price'].sum()
print(sum_)
# std
std = df['Price'].std()
print(std)
```
